# Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA

def pca_loadings(df, target, n):
    """
    Performs PCA and displays the loadings of features with the largest variance 
    explained with regards to a target variable. 
    
    Parameters
    ----------
    df : pandas dataframe
    
    target : str
        your dependent variable
        
    n : integer
        number of top loadings to display
    """
    dfx = df.drop(target, axis=1)
    X = np.array(dfx)
    pca = PCA(n_components=1)
    pca.fit(X)
    
    loadings = pd.DataFrame([({key: value for key,value in zip(dfx.columns, pca.components_[0])})],
             index=['weight']).transpose()
    loadings['abs_weight'] = np.abs(loadings['weight'])
    loadings.sort_values("abs_weight", ascending=False,inplace=True)
    top = loadings.head(n)
    return top
```

```python
pca_loadings(df, 'Target', 10)
```
